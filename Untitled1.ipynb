```python
df=pd.read_csv('us_states_covid19_daily.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
##removing colums with more than 50% null values
missing_percentage = df.isnull().sum() / len(df) * 100
threshold = 50
columns_to_remove = missing_percentage[missing_percentage > threshold].index
df = df.drop(columns=columns_to_remove)
df.head()
```

```python
print(df.duplicated().sum())
```

```python
df=df.drop_duplicates()
df=df.dropna()
```

```python
df.to_csv('python_cleaned us_states_covid19_daily.csv')
```
